```python
#Series de tiempo
# METODOS DEL TIEMPO
fecha=pd.to_datetime(pd.Timestamp.now()) # tiempo ahora
fecha

```

```python
# proxima clase
fin=pd.to_datetime(fecha+pd.Timedelta(weeks=1)) # fecha de una semana mas adelante comparado con el ahora
fin
```

```python
#Rango de fechas
fechas_1=pd.date_range(start=fecha, end=fin, periods=7)
fechas_1
```

```python
# periodo de dias 
fechas_2=pd.date_range(start=fecha,periods=8) # pongo 8 dias, con la fecha de inicio del ahora.
fechas_2
```

```python
fechas_3=pd.date_range(start=fecha,periods=8, freq='M')# PERIODO MESES
# dato el avance es relativo a meses, no a absoluto 
fechas_3
```

```python
mes_inicio=fecha.strftime('%Y-%m') # METODO QUE A LA FECHA LE EXTRAE EL AÑO Y EL MES
mes_inicio
```

```python
fechas_4=pd.period_range(start=mes_inicio, periods=8,freq='M')
fechas_4    
```

```python
#RESTAR FECHAS
cuanto_tiempo=fechas_3[7]-fechas_3[0]
cuanto_tiempo
```

```python
#cuanto tiempo en meses
cuanto_tiempo_meses=fechas_3[7].to_period('M')-fechas_3[0].to_period('M')
cuanto_tiempo_meses
```
