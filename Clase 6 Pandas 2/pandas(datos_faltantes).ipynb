```python
from numpy import nan, isnan
from pandas import read_csv
from sklearn.impute import SimpleImputer

url='https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv'
df=pd.read_csv(url3,sep=',',header=None)
df=df.replace(0,np.nan)
print(df.shape)
print(df.head())
```

```python
import sklearn

print(sklearn.__version__)
```

```python
# creo un arreglo en numpy con los valores
valores=df.values
# definir el imputador
imputador = SimpleImputer(missing_values=np.nan, strategy='mean')

# transformar el dataset

transformados=imputador.fit_transform(valores)
#Contar el numeros de Nan de cada columna

print("Missing: ", isnan(transformados).sum())
```

```python
transformados=pd.DataFrame(transformados)
transformados=transformados.rename(columns=
                                        {0:'NEMB',1:'GLU',2:'PART',3:'GROSS',4:'HUR',5:'BMI',6:'FPRED',7:'AGE',8:'CLASS'})
print(transformados.head().round(2))
```

```python
imputador=SimpleImputer(missing_values=-999,strategy='mean') # valores a buscar -999 y estrategia
                                                             # cuando los escuentre el median

# transformar el dataset
transformados=imputador.fit_transform(transformados)
# contar el numeros de nan de cada columna
print("Missing ", isnan(transformados).sum())
```
