```python
import pandas as pd
import numpy as np

df=pd.read_csv('./BTCUSD_1hr.csv', sep=',') # cargo el archivo, no olvidar de poner la ,
df.head()  # muestro las 5 primeras
```

```python
df.describe().round(1)
```

```python
print(df.isna().sum()) # busco los nulos, y si los hay los suma
```

```python
df['Date']
```

```python
a="2020-11-09 00:00:00"
a[5:7] # hago slicing entre el caracter 5 a 7
```

```python
def extraer_mes(x): # funcion para traer el numero que hace referencia al mes
    return x[5:7]
```

```python
def extraer_ano(x): # funcion que hace referencia al año
    return x[0:4]
```

```python
df["mes"]=df['Date'].apply(extraer_mes) # al extraer año y mes, agrego las columnas al df poriginal
                                        # que me va a servir para el group by 
df["año"]=df['Date'].apply(extraer_ano)
df.head()
```

```python
df.groupby(["año","mes"])[['Open','High','Low','Close','Volume']].mean().round(1)
# agrupo por año y por mes, mostrando las variable y redondeo. mostrando la media
```
