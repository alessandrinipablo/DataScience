```python
import pandas as pd
# creo las listas
frutas=['manzana','naranaja','banana','uva']
precio=[1.50,2.50,3.50,4]
cantidad=[2,4,5,9]
# luego creo el DataFrame(df)= invoco pandas.Dataframe y como parametro el paso un diccionario
# donde la clave le pongo el nombre de la columna, y en el valor , los valores que en este caso 
# son las frutas, precios, cantidad.
df=pd.DataFrame({'P':precio,'C':cantidad,'F':frutas})
df
```

```python
# uso la lista del los nombre de las frutas como index
frutas=['manzana','naranaja','banana','uva']
precio=[1.50,2.50,3.50,4]
cantidad=[2,4,5,9]
# uso las fruta como index
df=pd.DataFrame({'P':precio,'C':cantidad},index=frutas)
df
```

```python
# acceder a los elemntos de los dataframes 
#usado loc y iloc
# pongo df.iloc y entre corchetes el indice al cual quiero acceder
df.iloc[2]
```

```python
# usando loc ( accede por el nombre , no por posicion)
df.loc["uva"]
```

```python
# RUTA DEL ARCHIVO
file_path = './pune_1965_to_2002.csv'  # direccion relativa
# file_path= 'C:/Users/Usuario/Desktop/Data Science/Clase 6 Pandas 2/pune_1965_to_2002.csv' # direccion absoluta
# leo el archivo csv y le paso el parametro que es la direccion
df=pd.read_csv(file_path)
```

```python
# creo el dataFrame lluvias que es igual al que abri
df_lluvias_archivo=df
# MUESTRO LAS 5 PRIMERAS FILAS SI NO LE PASO NI UN NUMERO.
df_lluvias_archivo.head()
```

```python
# Dejar Year como index
df_lluvias_archivo.set_index('Year',inplace=True)  # seteo el index por defecto, y ahora lo dejo que el mismo sea el Año
df_lluvia=df_lluvias_archivo
df_lluvia.head()
```

```python
df_lluvia.sum() # suma por columnas son funciones de agregacion
```

```python
df_lluvia.mean() # promedio de lluvia por mes
```

```python
# puedo sumar horizontalmente ahora 
df_lluvia.sum(axis='columns')
# df_lluvia.sum()
```

```python
df_lluvia.sum(axis=0) # 0 sumaria por filas
#df_lluvia.sum(axis=1) # 1 sumaria por columnas
```

```python
df_lluvia.T.mean() # traspone los valores y saca la media
```

```python
#df_lluvia.describe().round(1) # aca hay otro ejemplo de como es la trasposicion,.
df_lluvia.T.describe().round(1)
```

```python
#METODO GROUP BY
print(df_lluvias_archivo.groupby('Year')['Jan','Feb','Mar'].sum().head(15))
```
