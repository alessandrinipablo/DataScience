```python
# Lista de marcas de autos
cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES']

# Datos de ventas de cada marca
data = [23, 17, 35, 29, 12, 41]

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(8, 7))
# Cambia el fondo del grfico, por defecto es blanco, se puede usar directamente grey
fig.set_facecolor('#b2b2b2')   

# Crear el grfico de pie
ax.pie(data, labels=cars, radius=0.7, startangle=90, explode=(0, 0, 0, 0, 0, 0.1), autopct="%1.1f%%", colors=sns.color_palette("hls", 7))

# Agregar una leyenda en la esquina inferior derecha
plt.legend(loc="lower right", labels=cars, fontsize=10)

# Mostrar el grfico
plt.show()
```

IMPORTANTE
#b2b2b2: Es un valor hexadecimal que representa un color en el modelo de colores RGB (rojo, verde y azul). En este caso, representa un gris medio.
labels: Es una lista con los nombres de las etiquetas que se mostrarn en el grfico.
radius: Es el radio del grfico circular, determina el tamao del grfico.
startangle: Es el ngulo inicial de la primera divisin del grfico, determina la posicin inicial de las divisiones.
explode: Es una lista que especifica la separacin de cada divisin del grfico circular respecto al centro. Permite resaltar una porcin del grfico.
autopct="%1.1f%%": Especifica el formato de etiqueta para mostrar el valor porcentual en cada porcin del grfico.
colors: Es una variable que almacena una paleta de colores generada con la funcin color_palette() del paquete Seaborn en Python. Se utiliza para asignar colores a las porciones del grfico.
Estos elementos se combinan para crear un grfico de pie con etiquetas, colores personalizados, tamao y estilo visual.

```python
# Crear la figura y los ejes con tamao personalizado

fig, ax = plt.subplots(figsize=(14, 3)) 

# Graficar las precipitaciones de enero y febrero
ax.plot(df_lluvia.index, df_lluvia['Jan'], label='Precipitaciones de enero')
ax.plot(df_lluvia.index, df_lluvia['Feb'], label='Precipitaciones de febrero', color='C1')

# Obtener los valores mximos de enero y febrero
maximo_enero = df_lluvia['Jan'].max()
maximo_febrero = df_lluvia['Feb'].max()

# Graficar lneas horizontales para los mximos de enero y febrero
ax.axhline(maximo_enero, color='red', linestyle='--', alpha=0.5, linewidth=3, label='Mxima de enero')
ax.axhline(maximo_febrero, color='red', linestyle=':', alpha=0.5, linewidth=3, label='Mxima de febrero')

# Establecer etiquetas de los ejes y ttulo del grfico
ax.set_xlabel('Ao')
ax.set_ylabel('Precipitacin (mm.)')
ax.set_title('Precipitaciones de enero y febrero')

# Establecer los lmites del eje x
ax.set_xlim(df_lluvia.index[0], df_lluvia.index[-1])

# Agregar leyenda al grfico con tamao de fuente personalizado
ax.legend(fontsize=8)

# Establecer el tamao de las etiquetas de los ejes
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=6)

# Mostrar el grfico
plt.show()
```
