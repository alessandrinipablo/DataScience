```python
import matplotlib.pyplot as plt
import numpy as np

#Generamos una muestra de 5000 elementos con media 0 y desviación estándar 1
normal_samples = np.random.normal(0, 1, size=5000)

#Graficamos nuestros datos
fig, ax = plt.subplots()
plt.hist(normal_samples, bins=50, density=True)
plt.title('Distribución normal', size=14)
plt.xlabel('Muestras')
plt.ylabel('Densidad')

plt.show()
```

```python
df.describe()
```

```python
df.describe(include='category')
```

```python
# el método value_counts() permite contar las frecuencias de cada categoria
df['cylinders'].value_counts()
```

```python
# si utilizamos el parametro normalize se obtienen los datos
# con frecuencia relativa o porcentaje
df['cylinders'].value_counts(normalize=True)
```

```python
# el ID de los autos esta almacenado en el index, lo vamos a extraer y renombrar para poder comprobar su cardinalidad
df.reset_index(inplace=True)
df.rename(columns={'index':'ID'}, inplace=True)

# Para evaluar la cardinalidad de las variables podemos utilizar nunique()
# Este método cuenta los valores únicos de una variable
df.nunique()
```

```python
# Tambien se puede utilizar en una columna única
df.origin.nunique()
```

```python
# y tl método unique() muestra los datos únicos
df.origin.unique()
```

```python
promedio = df.acceleration.mean() # promedio o media
mediana = df.acceleration.median() # mediana o Q2
moda = df.acceleration.mode() # moda
primer_cuartil = df.acceleration.quantile(0.25) # Q1 cuartíl 1 --> percentíl 25
tercer_cuartil = df.acceleration.quantile(0.75) # Q3 cuartíl 2 --> percentíl 75
percentil_99 = df.acceleration.quantile(0.99) # percentíl 99


# Imprimir los resultados
print("Promedio: ", promedio)
print("Mediana: ", mediana)
print("Moda: ", moda)
print("Primer cuartil: ", primer_cuartil)
print("Tercer cuartil: ", tercer_cuartil)
print("Percentil 99: ", percentil_99)
```
